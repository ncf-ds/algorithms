Visualizations of `dual pivot quicksort` (two versions) and `introsort`

```python
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
import datetime
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import seaborn as sns

%matplotlib inline
sns.set(style="darkgrid")
mpl.rc("figure", figsize=(12, 6))
```

### Plots

```python
df1 = pd.read_csv('./data/20150920-dual-pivot-quicksort-analysis-2.csv')
df2 = pd.read_csv('./data/20150920-dual-pivot-quicksort-analysis-3.csv')
df3 = pd.read_csv('./data/20150920-introsort-analysis.csv')
df1['algorithm'] = 'Dual Pivot 2'
df2['algorithm'] = 'Dual Pivot 3'
df3['algorithm'] = 'Introsort'
df3 = df3[df3.N <= 1000]
df = pd.concat([df1, df2, df3])
```

```python
g = sns.lmplot("N", "T (s)", data=df[df['b/N'] == 1], hue='algorithm', fit_reg=False, ci=None, palette="muted",
           scatter_kws={"s": 50, "alpha": 1})
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Raw Data, b/N = 1') 
```

```python
g = sns.lmplot(x="log(N)", y="log(T)", data=df[df['b/N'] == 1],  hue='algorithm', ci=None, palette="muted",
           scatter_kws={"s": 50, "alpha": 1})
plt.subplots_adjust(top=0.9)
g.fig.suptitle('log-log data with trendline') 
```

```python
### Slope calculations
dff = df.rename_axis({"log(N)": "y", "log(T)":"x"}, axis=1) #smf.olf Doesn't like these names for some reason.
dff = dff[dff['b/N'] == 1]
```

```python
slope_1 = smf.ols(formula="y ~ x", data=dff[dff.algorithm == 'Dual Pivot 2']).fit().params["x"]
slope_2 = smf.ols(formula="y ~ x", data=dff[dff.algorithm == 'Dual Pivot 3']).fit().params["x"]
slope_3 = smf.ols(formula="y ~ x", data=dff[dff.algorithm == 'Introsort']).fit().params["x"]
```

```python
print("Dual Pivot 2 Slope: " + str(slope_1))
print("Dual Pivot 3 Slope: " + str(slope_2))
print("Introsort Slope: " + str(slope_3))
```

```python
g = sns.lmplot("N", "ncomparisons", data=df[df["b/N"] == 1], hue='algorithm', fit_reg=False, ci=None, palette="muted",
           scatter_kws={"s": 50, "alpha": 1})
plt.subplots_adjust(top=0.9)
g.fig.suptitle('N vs. Number of Comparisons,  b/N = 1') 
```

```python
g = sns.lmplot("N", "nswaps", data=df[df["b/N"] == 1], hue='algorithm', fit_reg=False, ci=None, palette="muted",
           scatter_kws={"s": 50, "alpha": 1})
plt.subplots_adjust(top=0.9)
g.fig.suptitle('N vs. Number of Swaps, b/N = 1') 
```

```python
#%matplotlib qt
with sns.axes_style("ticks"):
    fig = sns.plt.figure(figsize=(8,8))
    #fig.figure(figsize=(8,8))
    ax = fig.add_subplot(111, projection='3d')
    
    m = 'o'
    xs = df["N"][df.algorithm == 'Dual Pivot 2']
    ys = df["T (s)"][df.algorithm == 'Dual Pivot 2']
    zs = df["b/N"][df.algorithm == 'Dual Pivot 2']
    ax.scatter(xs, ys, zs, c='b', marker=m, label='Dual Pivot 2')
    
    xs2 = df["N"][df.algorithm == 'Dual Pivot 3']
    ys2 = df["T (s)"][df.algorithm == 'Dual Pivot 3']
    zs2 = df["b/N"][df.algorithm == 'Dual Pivot 3']
    ax.scatter(xs2, ys2, zs2, c='g', marker=m, label='Dual Pivot 3')
    
    xs3 = df["N"][df.algorithm == 'Introsort']
    ys3 = df["T (s)"][df.algorithm == 'Introsort']
    zs3 = df["b/N"][df.algorithm == 'Introsort']
    ax.scatter(xs3, ys3, zs3, c='r', marker=m, label='Introsort')
    
    ax.legend()
    ax.set_xlabel('N')
    ax.set_ylabel('T (s)')
    ax.set_zlabel('b/N')
    ax.title.set_text('T vs N vs b/N')
    plt.show()
```
